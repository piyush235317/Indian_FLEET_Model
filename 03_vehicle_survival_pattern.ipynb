```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define retirement years
public_retirement_year, commercial_retirement_year = 20, 15
simulation_years = 30

# Step function to simulate vehicle status
def step_function(active_years, retirement_year):
    status = [1] * active_years + [0] * (retirement_year - active_years) + [0] * (simulation_years - retirement_year)
    return status

# Simulate survival patterns
public_survival_pattern = step_function(public_retirement_year, public_retirement_year)
commercial_survival_pattern = step_function(commercial_retirement_year, commercial_retirement_year)

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

plt.plot(public_survival_pattern, label="Public Vehicles", color="tab:blue", linewidth=3)
plt.plot(commercial_survival_pattern, label="Commercial Vehicles", color="tab:orange", linewidth=3)

plt.axvline(x=public_retirement_year, linestyle="--", color="tab:blue", alpha=0.7)
plt.axvline(x=commercial_retirement_year, linestyle="--", color="tab:orange", alpha=0.7)

# Labels and annotations
plt.xlabel("Years", fontsize=14)
plt.ylabel("Vehicle Status", fontsize=14)
plt.title("Survival Patterns of Public and Commercial Vehicles in India", fontsize=16)
plt.legend(fontsize=12)

plt.xticks(fontsize=12)
plt.yticks([0, 1], ['Retired', 'Active'], fontsize=12)

plt.grid(True, linestyle="--", alpha=0.5)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

arrow_color = "black"

plt.annotate("Compulsory Retirement\nfor Public Vehicles", xy=(public_retirement_year, 1), xytext=(public_retirement_year+5, 0.8),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2", color=arrow_color), fontsize=10, color="tab:blue",
             bbox=dict(boxstyle="round,pad=0.4", fc="white", ec="tab:blue"))

plt.annotate("Compulsory Retirement\nfor Commercial Vehicles", xy=(commercial_retirement_year, 1), xytext=(commercial_retirement_year+2, 0.8),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2", color=arrow_color), fontsize=10, color="tab:orange",
             bbox=dict(boxstyle="round,pad=0.4", fc="white", ec="tab:orange"))

plt.savefig('life_cycle_assessment.png', dpi=300)

plt.tight_layout()
plt.show()
```
